### 图片颜色替换  
- 读取图片RGB -> 转换成HSV格式 -> 生成想要替换的颜色的掩膜mask -> 遍历图像np数组替换掩膜位置的颜色
- 图片复杂时，需要在制作掩膜后进行开运算，来去除干扰

```python
# 导入
import cv2
import numpy as np

# 1.读入图像
image_np = cv2.imread("../image/four_color.png")

# 2.转换为HSV模式
image_hsv = cv2.cvtColor(image_np, cv2.COLOR_BGR2HSV)

# 3.制作掩膜
# 定义上下限
red_low1 = np.array([0, 43, 46])
red_low2 = np.array([156, 43, 46])
red_high1 = np.array([10, 255, 255])
red_high2 = np.array([180, 255, 255])
# 求第一个范围的掩膜
image_mask1 = cv2.inRange(image_hsv, red_low1, red_high1)
# 求第二个范围的掩膜
image_mask2 = cv2.inRange(image_hsv, red_low2, red_high2)
# 将两个掩膜想与，求出总掩膜
image_mask = cv2.bitwise_or(image_mask1, image_mask2)


# 4.开运算
# 定义核大小
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
# 对掩膜进行开运算
# 掩膜虽然能标出目标区域，但往往含有噪声、孔洞、毛边或断裂点。 对其进行开运算就可以消除这些噪声
image_mask_open = cv2.morphologyEx(image_mask, cv2.MORPH_OPEN, kernel)

# 5.图片颜色替换
# 核心就是image_np 的重新赋值
# for i in range(image_mask_open.shape[0]):
#     for j in range(image_mask_open.shape[1]):
#         if image_mask_open[i, j] == 255:
#             #image_np[i, j] = (255, 0, 0)
#             image_np[i, j] = (123, 32, 65)
# 快速代码
image_np[image_mask_open == 255] = (255, 0, 0)

# 6.显示结果

cv2.imshow("image_mask_open", image_mask_open)
cv2.imshow("image_np", image_np)
cv2.waitKey()
```
