# 形态变换学
- 卷积核的生效范围是：卷积区域内核值为1的部分  
     例如：十字核，他是3*3，但只有核心十字形状的部分为1
- 腐蚀：在卷积核的生效范围内找最小值
- 膨胀：在卷积核的生效范围内找最大值

```python
# 对二值化图像进行腐蚀操作

# 先导入opencv的库，方便后面直接调用函数
import cv2

# 1.读取一张要操作的彩色图像
image_np = cv2.imread("../image/image.jpg")

# 2.对图像进行灰度化
image_gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

# 3.对图像进行二值化
ret, image_thresh = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

# 4.进行腐蚀操作：
# 4.1 构建腐蚀操作所需要用到的核
# 第一个参数：指定核的形状  十字形状 cv2.MORPH_CROSS 矩形 cv2.MORPH_RECT 椭圆 cv2.MORPH_ELLIPSE
# 第二个参数：
kernel = cv2.getStructuringElement(
    cv2.MORPH_CROSS, (3, 3), anchor=(1, 1)
)  # anchor=(1,1)选择核的中心
# 4.2 进行滑动计算
# 腐蚀
image_erode = cv2.erode(image_thresh, kernel)

# 5.显示图像
cv2.imshow("image_thresh", image_thresh)
cv2.imshow("image_erode", image_erode)
cv2.waitKey()
```

```python
# 对二值化图像进行膨胀操作

# 先导入opencv的库，方便后面直接调用函数
import cv2

# 1.读取一张要操作的彩色图像
image_np = cv2.imread("../image/image.jpg")

# 2.对图像进行灰度化
image_gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

# 3.对图像进行二值化
ret, image_thresh = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

# 4.进行腐蚀操作：
# 4.1 构建腐蚀操作所需要用到的核
# 第一个参数：指定核的形状  十字形状 cv2.MORPH_CROSS 矩形 cv2.MORPH_RECT 椭圆 cv2.MORPH_ELLIPSE
# 第二个参数：
kernel = cv2.getStructuringElement(
    cv2.MORPH_CROSS, (3, 3), anchor=(1, 1)
)  # anchor=(1,1)选择核的中心
# 4.2 进行滑动计算
# 腐蚀
image_dilate = cv2.dilate(image_thresh, kernel)

# 5.显示图像
cv2.imshow("image_thresh", image_thresh)
cv2.imshow("image_dilate", image_dilate)
cv2.waitKey()
```
