```python
# 导入
import cv2

# 1.读取原始图像和模板图
# 原图
image_y_np = cv2.imread("../image/image.jpg")
# 模板图
image_j_np = cv2.imread("../image/jinbi.png")

# 截取原始图像的numpy
image_j_np_rows, image_j_np_cols = image_j_np.shape[:2]

# 截取图像的ROI区域
roi = image_y_np[:image_j_np_rows, :image_j_np_cols]

# 2.对模板进行灰度化和二值化

image_j_np_gray = cv2.cvtColor(image_j_np, cv2.COLOR_BGR2GRAY)

# 进行的是反阈值化，将前景变为黑色
ret, image_j_np_thresh = cv2.threshold(image_j_np_gray, 127, 255, cv2.THRESH_BINARY_INV)

# 3.进行与运算
# src1、src2：要进行按位与运算的两张图像（大小和类型必须相同）
# mask（可选）：掩膜图像，只在 mask 不为 0 的位置才会计算结果，其他位置结果为 0
# 这里 mask 的作用是：
# 只在掩膜中 非零（白色）区域 执行 AND 操作，其他地方输出为 0。
roi_and = cv2.bitwise_and(roi, roi, mask=image_j_np_thresh)

# 4.图像融合
# 使用add函数
roi_add = cv2.add(roi_and, image_j_np)

# 5.将融合后的图像重新赋值刀原始图像中
image_y_np[:image_j_np_rows, :image_j_np_cols] = roi_add

# 6.显示结果
cv2.imshow("image_y_np", image_y_np)
cv2.imshow("image_j_np_thresh", image_j_np_thresh)
cv2.imshow("roi_and", roi_and)
cv2.waitKey(0)
```
