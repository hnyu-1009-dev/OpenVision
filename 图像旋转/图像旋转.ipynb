### 图像旋转需要用到一下核心技术：
- 1.旋转矩阵：旋转+平移
- 2.插值方法：1.最近邻  2.双线性（本块知识点用到的最多）  3.像素区域 。。。。。。。
- 3.边界填充
- 4.图片旋转，逆时针方向位正方向

```python
# 对图像进行认识角度的旋转
# 在OpenCV中逆时针方向为正

# 导入
import cv2

# 1.读取一张图片
image_np = cv2.imread("../image/image.jpg")
image_shape = image_np.shape

# 定义旋转的方向和角度 以及 图片缩放的比例
angle = 4.5
scale = 0.5

# 定义旋转矩阵
# cv2.getRotationMatrix2D： 构建一个2*3的旋转矩阵，包含了图像的旋转，平移与缩放
# 第一个参数：旋转中心
# 第二个参数：旋转角度
# 第三个参数：缩放比例
M = cv2.getRotationMatrix2D((image_shape[1] // 2, image_shape[0] // 2), angle, scale)

# 有了旋转矩阵之后，可以进行图像旋转
# cv2.warpAffine：对图像进行旋转
# 第一个参数：要旋转的图像
# 第二个参数：通过cv2.getRotationMatrix2D获得到的旋转矩阵
# 第三个参数：输出图像的大小
image_rotation = cv2.warpAffine(
    image_np,
    M,
    (image_shape[1], image_shape[0]),
    flags=cv2.INTER_LINEAR,
    borderMode=cv2.BORDER_CONSTANT,
    borderValue=(0, 0, 0),
)

# 显示旋转的结果
cv2.imshow("image_rotation", image_rotation)
cv2.waitKey()
```
