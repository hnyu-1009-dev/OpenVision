```python
# 对图像进行反转操作
# 导入
import cv2

# 1.读取图片
image_np = cv2.imread("../image/image.jpg")

# 2. 使用flip函数对图像进行镜像的反转
# cv2.flip:对图像进行镜像翻转
# 第一个参数：要反转的原始图像
# 第二个参数：标志位，0：表示绕x轴进行上下反转  >0:表示绕y轴进行左右反转  <0:表示绕x轴和y轴各进行一次反转
iamge_flip = cv2.flip(image_np, 0)
# 3.输出
cv2.imshow("image_np", image_np)
cv2.imshow("iamge_flip", iamge_flip)
cv2.waitKey(0)
```
