```python
# 
import cv2

# 1.
image_np = cv2.imread("../image/image.jpg")

# 
height, width = image_np.shape[0], image_np.shape[1]
print(height, width)

try:
    # 2.
    # 
    x_min, x_max = 60, 160
    y_min, y_max = 60, 160

    # 
    if not (x_min >= 0 and x_max <= width and y_min >= 0 and y_max <= height):
        # raise 
        raise OverflowError("x_min or x_max or y_min or y_max overflower!!")

    # cv2.rectangle 
    cv2.rectangle(
        image_np, (x_min - 2, y_min - 2), (x_max + 2, y_max + 2), (0, 0, 255), 2
    )

    # 3.np
    image_roi = image_np[y_min:y_max, x_min:x_max]

    # 4.
    cv2.imshow("image_np", image_np)
    cv2.imshow("image_roi", image_roi)
    cv2.waitKey()


except Exception as e:
    print(e)
```
