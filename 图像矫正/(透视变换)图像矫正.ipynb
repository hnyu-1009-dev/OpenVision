```python
# 导入
import cv2
import numpy as np

# 1.读取图片
image_np = cv2.imread("../image/jiaozheng.png")

# 获取图像的大小
image_shape = image_np.shape

# 2.定义原始图像中的四个顶点坐标
points = np.float32([[118, 60], [350, 90], [75, 195], [338, 230]])
# 定义目标图像中，这四个顶点坐标所对应的位置
points2 = np.float32(
    [[0, 0], [image_shape[1], 0], [0, image_shape[0]], [image_shape[1], image_shape[0]]]
)

cv2.line(image_np, [118, 60], [350, 90], (0, 0, 255), 1, lineType=cv2.LINE_AA)
cv2.line(image_np, [75, 195], [338, 230], (0, 0, 255), 1, lineType=cv2.LINE_AA)
# 3.获取透视变换矩阵
M = cv2.getPerspectiveTransform(points, points2)

# 4.进行透视变换
image_warpPerspective = cv2.warpPerspective(
    image_np, M, (image_shape[1], image_shape[0])
)

# 5.显示图像
cv2.imshow("image_np", image_np)
cv2.imshow("image_warpPerspective", image_warpPerspective)
cv2.waitKey()
```
