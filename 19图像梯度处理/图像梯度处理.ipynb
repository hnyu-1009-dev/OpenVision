```python
import cv2
import numpy as np

# ==============================
# 1️⃣ 读取并检查图像
# ==============================

# 读取图像（默认以 BGR 格式）
# cv2.imread(filename, flags)
#   - filename：图像路径
#   - flags：读取方式
#       cv2.IMREAD_COLOR（默认）→ 彩色图
#       cv2.IMREAD_GRAYSCALE → 灰度图
# 返回值：numpy.ndarray 类型图像矩阵
image_path = "../image/my_shudu.jpg"
image_np = cv2.imread(image_path)

# 检查图像是否加载成功
if image_np is None:
    raise FileNotFoundError(f"⚠️ 图像未找到，请检查路径是否正确: {image_path}")

# 转为灰度图，减少计算量
# cv2.cvtColor(src, code)
#   - src：输入图像
#   - code：颜色空间转换标识，如：
#       cv2.COLOR_BGR2GRAY（BGR → 灰度）
#       cv2.COLOR_BGR2RGB（BGR → RGB）
# 返回值：单通道灰度图
image_np_gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

# ==============================
# 2️⃣ Sobel 边缘检测
# ==============================

# Sobel 算子用于计算图像的梯度（灰度变化率）
# cv2.Sobel(src, ddepth, dx, dy, ksize, scale, delta, borderType)
#   - src：输入图像（通常为灰度图）
#   - ddepth：输出图像深度（如 cv2.CV_16S 防止溢出）
#   - dx：X 方向导数阶数（1 表示一阶导）
#   - dy：Y 方向导数阶数（1 表示一阶导）
#   - ksize：卷积核大小（1、3、5、7）奇数，越大平滑越强
#   - scale（默认=1）：可选缩放因子
#   - delta（默认=0）：偏移量（加到结果上）
#   - borderType：边界处理方式（默认 cv2.BORDER_DEFAULT）
# 返回值：包含梯度值的图像矩阵（可能包含负值）
sobel_x = cv2.Sobel(
    image_np_gray,  # 输入灰度图
    cv2.CV_16S,  # 输出类型，16位有符号，防止负值截断
    1,  # dx=1 → 计算 X 方向梯度（垂直边缘）
    0,  # dy=0 → 不计算 Y 方向导数
    ksize=3,  # 卷积核大小 3x3
)

sobel_y = cv2.Sobel(
    image_np_gray,
    cv2.CV_16S,
    0,  # dx=0 → 不计算 X 方向导数
    1,  # dy=1 → 计算 Y 方向梯度（水平边缘）
    ksize=3,
)
sobel_center = cv2.Sobel(
    image_np_gray,
    cv2.CV_16S,
    1,  # dx=0 → 不计算 X 方向导数
    1,  # dy=1 → 计算 Y 方向梯度（水平边缘）
    ksize=3,
)
# 将 16 位带符号数据转换为 8 位绝对值图像（便于显示）
# cv2.convertScaleAbs(src)
#   - src：输入图像
#   - 返回：uint8 图像（0-255）
# 该函数等价于：dst = np.abs(src) 并归一化到 [0,255]
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
sobel_center = cv2.convertScaleAbs(sobel_center)

# 合并两个方向的梯度
# cv2.addWeighted(src1, alpha, src2, beta, gamma)
#   - src1, src2：输入图像
#   - alpha, beta：加权系数（一般取 0.5）
#   - gamma：常数偏移量
# 返回值：加权融合后的图像
sobel_xy = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

# ==============================
# 3️⃣ 可视化结果
# ==============================

# 显示图像窗口
# cv2.imshow(windowName, image)
#   - windowName：窗口标题
#   - image：要显示的图像矩阵
cv2.imshow("Original Image", image_np)
cv2.imshow("Gray Image", image_np_gray)
cv2.imshow("Sobel X (Vertical Edges)", sobel_x)
cv2.imshow("Sobel Y (Horizontal Edges)", sobel_y)
cv2.imshow("Sobel Y (Horizontal Edges)", sobel_y)
cv2.imshow("Sobel center (Horizontal Edges)", sobel_center)
cv2.imshow("Sobel XY (Combined)", sobel_xy)

# 等待按键事件
# cv2.waitKey(delay)
#   - delay：等待毫秒数；0 表示无限等待
cv2.waitKey(0)

# 销毁所有窗口
# cv2.destroyAllWindows()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# ==============================
# 1️⃣ 读取灰度图
# ==============================
image = cv2.imread("../image/my_shudu.jpg")
if image is None:
    raise FileNotFoundError("⚠️ 图像未找到，请检查路径！")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# ==============================
# 2️⃣ Laplacian 算子
# ==============================
# cv2.Laplacian(src, ddepth, ksize, scale, delta, borderType)
#   - src：输入灰度图像
#   - ddepth：输出图像深度（推荐 cv2.CV_16S）
#   - ksize：卷积核大小（1, 3, 5，必须为奇数）
#   - scale：可选缩放因子（默认1）
#   - delta：偏移量（默认0）
#   - borderType：边界处理方式
# 功能：计算图像灰度的二阶导，检测灰度变化剧烈的边缘
laplacian = cv2.Laplacian(gray, cv2.CV_16S, ksize=3)

# 取绝对值并转为8位图
laplacian = cv2.convertScaleAbs(laplacian)

# ==============================
# 3️⃣ 显示结果
# ==============================
cv2.imshow("Original", image)
cv2.imshow("Laplacian (Edge Detection)", laplacian)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
