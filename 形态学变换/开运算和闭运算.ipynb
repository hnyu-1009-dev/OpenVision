```python
import cv2
import numpy as np


# 读取图像并转为灰度
iamge = cv2.imread('../image/image.jpg', 0)


# 二值化
_, binary = cv2.threshold(iamge, 127, 255, cv2.THRESH_BINARY)


# 定义结构元素
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))


# 开运算：先腐蚀后膨胀
open_result = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)


# 显示结果
cv2.imshow('Original', binary)
cv2.imshow('Open Operation', open_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


# 读取图像并转为灰度
iamge = cv2.imread('../image/image.jpg', 0)


# 二值化
_, binary = cv2.threshold(iamge, 127, 255, cv2.THRESH_BINARY)


# 定义结构元素
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))


# 闭运算：先膨胀后腐蚀
close_result = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)


# 显示结果
cv2.imshow('Original', binary)
cv2.imshow('Close Operation', close_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
